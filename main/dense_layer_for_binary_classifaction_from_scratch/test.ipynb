```python
import numpy as np
from neuron5 import Neuron

# Load data
lib_train = np.load('mnist_binary_01.npz')
X_3D, Y = lib_train['x_train'], lib_train['y_train']

# Reshape and normalize input data
X = X_3D.reshape((X_3D.shape[0], -1))
X = X / 255.0  # Normalize pixel values to [0, 1]

# Ensure Y has proper shape
if Y.ndim == 1:
    Y = Y.reshape(1, -1)


np.random.seed(0)
neuron = Neuron(X.shape[1])
A = neuron.forward_prop(X)
cost = neuron.cost(Y, A)
# Clip A values to prevent log(0) or log(1)
A = np.clip(A, 1e-15, 1-1e-15)
predications = neuron.predict(X, Y)
neuron.back_prop(X, Y, A, 0.5)
print(neuron.w)
print(neuron.b)
print(predications)
print(cost)

print(cost)
```

```python
np.random.seed(0)
neuron = Neuron(X.shape[1])
A, cost = neuron.train(X, Y, iterations=10)
accuracy = np.sum(A == Y) / Y.shape[1] * 100
print("Train cost:", np.round(cost, decimals=10))
print("Train accuracy: {}%".format(np.round(accuracy, decimals=10)))
print("Train data:", np.round(A, decimals=10))
print("Train Neuron A:", np.round(neuron.A, decimals=10))


A, cost = neuron.predict(X, Y)
accuracy = np.sum(A == Y) / Y.shape[1] * 100
print("Dev cost:", np.round(cost, decimals=10))
print("Dev accuracy: {}%".format(np.round(accuracy, decimals=10)))
print("Dev data:", np.round(A, decimals=10))
print("Dev Neuron A:", np.round(neuron.A, decimals=10))
```
